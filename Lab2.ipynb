```python
import pylab
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
log_df = pd.read_csv("wc_day6_1_sample.csv",
                    names = ['ClientID','Date','Time','URL','ResponseCode','Size'],
                    na_values = '-')
```

```python
#Question 1
multi_grouped = log_df.groupby(['Date','ResponseCode'])
multi_grouped.get_group(('30/Apr/1998',404)).count().Size
```

```python
#Question 2
may_df = log_df[log_df['Date'] == '01/May/1998']
may_df['DateTime'] = pd.to_datetime(may_df.apply(lambda row: row['Date'] + ' ' + row['Time'], axis = 1)) 
hour_grouped= may_df.groupby(lambda x: may_df['DateTime'][x].hour)
```

```python
ax = hour_grouped['ClientID'].nunique().plot(kind = 'bar')
ax.set_xlabel("Number of users")
ax.set_ylabel("Hours of the day")
ax.set_title("Number of users change in a day")
```

```python
#Question 3
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
x = hour_grouped.size().index
ax1.plot(x,hour_grouped['ClientID'].nunique(),'g-')
ax2.plot(x,hour_grouped['Size'].sum(),'r-')
 
ax1.set_xlabel("Hours of the day")
ax1.set_ylabel("Number of users", color='g')
ax2.set_ylabel("Size of the traffic handled", color = 'r')
```

```python
#Question 5     
log_df = pd.read_csv("wc_day91_1.csv",
                    names = ['ClientID','Date','Time','URL','ResponseCode','Size'],
                    na_values = '-')
```

```python
#Part I  Jul/24
Jul24_df = log_df[log_df['Date'] == '24/Jul/1998']
Jul24_df['DateTime'] = pd.to_datetime(Jul24_df.apply(lambda row: row['Date'] + ' ' + row['Time'], axis = 1)) 
hour_grouped= Jul24_df.groupby(lambda x: Jul24_df['DateTime'][x].hour)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
x = hour_grouped.size().index
ax1.plot(x,hour_grouped['ClientID'].nunique(),'g-')
ax2.plot(x,hour_grouped['Size'].sum(),'r-')
 
ax1.set_xlabel("Hours of the day")
ax1.set_ylabel("Number of users", color='g')
ax2.set_ylabel("Size of the traffic handled", color = 'r')
```

```python
log_client = Jul24_df.ClientID.unique()
randomID = np.random.choice(log_client,100)
sample = Jul24_df[Jul24_df['ClientID'].isin(randomID)]
sample['Hour'] = pd.DatetimeIndex(sample['DateTime']).hour
sample.plot(kind='scatter', x='ClientID', y='Hour')
```

```python
#Part II  Jul/25
Jul25_df = log_df[log_df['Date'] == '25/Jul/1998']
Jul25_df['DateTime'] = pd.to_datetime(Jul25_df.apply(lambda row: row['Date'] + ' ' + row['Time'], axis = 1)) 
hour_grouped= Jul25_df.groupby(lambda x: Jul25_df['DateTime'][x].hour)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
x = hour_grouped.size().index
ax1.plot(x,hour_grouped['ClientID'].nunique(),'g-')
ax2.plot(x,hour_grouped['Size'].sum(),'r-')
 
ax1.set_xlabel("Hours of the day")
ax1.set_ylabel("Number of users", color='g')
ax1.set_ylabel("Size of the traffic handled", color = 'r')
```

```python
log_client = Jul25_df.ClientID.unique()
randomID = np.random.choice(log_client,100)
sample = Jul25_df[Jul25_df['ClientID'].isin(randomID)]
sample['Hour'] = pd.DatetimeIndex(sample['DateTime']).hour
```

```python
sample.plot(kind='scatter', x='Hour', y='ClientID')
```

These three days show us a lot of differences and similars.
1. For all three days, we can find a kind of positive correlation between the number od users and the size of traffic.
2. However, the hours that people sent requests are different. For example, people only send requests at 22:00 and 23:00 on Jul 24th. 
3. comparing clientID with the hours, we also can find some interesting conclusion. At the beginning of the World Cup (01/May), people with less clientID loved to send requests all day, however people who had larger clientID loved to send requests at night. Some days later(25/Jul and 24/Jul), people with cilentID between 700000 and 1200000 send less requests than other people. And for Jul 25th, there is no patricular relation between hours and users who sent requests.
